```python
from urllib import request
import io
```

```python
url_page = r'https://dsmdaviz.com/wp-content/uploads/2018/05/10.png'
response = request.urlopen(url_page)
data = response.read()
data[0:100]
```

```python
# If you just want to save that as a png, something like this should work:
newFile = open('sample.png', 'wb')
newFile.write(data)
```

```python
# If you want to do anything with the image, you'll need to process it a bit:

# Keep the bytes in "data" as a "pseudo-file" --you could just save this as a .png file as above and load from disk.
img = io.BytesIO(data)

#if you want to pass it as a base64 txt:
import base64
img_str = base64.b64encode(img.getvalue())

img_str[0:150]
```

```python
# If you want to work with the image
from PIL import Image

pil_im = Image.open(img)
```
